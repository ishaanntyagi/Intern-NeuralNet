```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X=iris.data
y=iris.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(
    X,y,random_state=42,test_size=0.3)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
class IrisDataset(Dataset): #Defining a class , also taking Dataset which means we are inheriting from the torch.utils.Datset from PyTorch ,
    #this means That we can now use the basic Functionality of the Torch Datsset and we can customize how this can be accessed 
    def __init__ (self,features,labels): # defing a constructor , gets called automatically 
        #self refers to the Current object/Instance Being Created 
        #features and labels Ykwim 
         
        self.X = torch.tensor(features, dtype=torch.float32) # turns your Feature Data into Tensors that torch can read 
        self.y = torch.tensor(labels, dtype=torch.long) #torch.tensor here also meant the same and means this will be converted to Long Format 
#splitting of Dataset , for features and target variables  
        
        
def __len__(self):
    return len(self.x) # finds the length of the Samples in the dataset (finds out how many are there) 
#if self.X has 150 rows then , it will return 150

def __getitem__(self, idx):  #tells python how to get an item , given its position and index
    #if the datsset[5] => idx = 5. 
    
    return self.X[idx], self.y[idx] # returns the data ( features and Labels) for the sample Position at idx 
# this method is used in pytorch to get batches of data during training.
    
        
```

```python
dataset = IrisDataset(X,y)
```
